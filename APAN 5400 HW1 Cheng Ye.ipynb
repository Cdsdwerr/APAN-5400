Question 1:
Read the CSV file into a Pandas data frame and drop records where company names are None values:
Print the count of records (rows) in the resulting data frame
Print the first 5 records of the data frame

```python
#1
import pandas as pd

a = pd.read_csv("C:/Users/yeche/Downloads/crunchbase_odm_orgs.csv") #reading csv file

a = a.dropna(subset=['name']) #drop na values in name variable

print('Number of records after data cleaning is:',len(a)) # print count

a.head(5) #print the first 5 records of the data frame
```

Question 2:
Create a data frame that contains only the records of USA-based companies whose name starts with "Ac":
Print the count of records (rows) in the resulting data frame
Print the first 5 records of the data frame

```python
#2

b = a[(a['country_code'] == 'USA') & (a['name'].str.startswith('Ac'))]

print('Number of records that contains only the records of USA-based companies whose name starts with Ac is:',len(b)) #print count of rows

b.head(5)# print the first 5 records of the DataFrame
```

Question 3:
Convert the data frame from the previous step into a list of JSON objects:
Print the count of JSON objects
Print the first 5 JSON objects
Write all JSON objects into a text (JSON string) file
Print the number of records in the resulting file

```python
#3
import json
c = b.to_dict('records')

print('The count of JSON objects is:',len(c)) #print count

print("count5:")
for json_obj in c[:5]:
    print(json_obj) #print count5

with open('output.json', 'w') as file:
    file.write(pd.Series(c).to_json(orient='records', lines=True))
# Write JSON objects to a text file


with open('output.json', 'r') as file:
    records = sum(1 for _ in file)
print(records) # Print the number of records in the resulting file
```

Question 4:
